```python
%%time

# Faz umas estatísticas com as bolas (ordena, maném resultado, soma, média e desvio padrão)
def jogos_ordenados(bola_1: int, bola_2: int, bola_3: int, bola_4: int, bola_5: int, bola_6: int,
          bola_7: int, bola_8: int, bola_9: int, bola_10: int, bola_11: int, bola_12: int, 
          bola_13: int, bola_14: int, bola_15: int, ) -> list:
    """Retorna os 15 numeros do resultado da lotofacil, ordenados"""
    return sorted([bola_1, bola_2, bola_3, bola_4, bola_5, bola_6, 
                   bola_7, bola_8, bola_9, bola_10, bola_11, bola_12,
                   bola_13, bola_14, bola_15])

def jogos_desordenados(bola_1: int, bola_2: int, bola_3: int, bola_4: int, bola_5: int, bola_6: int,
          bola_7: int, bola_8: int, bola_9: int, bola_10: int, bola_11: int, bola_12: int, 
          bola_13: int, bola_14: int, bola_15: int, ) -> list:
    """Retorna os 15 numeros do resultado da lotofacil, desordenados"""
    return  ([bola_1, bola_2, bola_3, bola_4, bola_5, bola_6, 
                   bola_7, bola_8, bola_9, bola_10, bola_11, bola_12,
                   bola_13, bola_14, bola_15])

def  soma_jogos(lista: list) -> int:
    """ retorna soma dos jogos """
    return sum(lista)

def  media_jogos(lista: list) -> int:
    """ retorna media dos jogos """
    return int(np.mean(lista))

def  desvio_padrao_jogos(lista: list) -> int:
    """ retorna desvio padrao dos jogos """
    return int(np.std(lista))

# Separa entre em categorias: Bolas entre 1 e 5, Bolas entre 6 e 10, Bolas entre 11 e 15, etc  
def entre_numeros(df: pd.DataFrame, primeiro: int, segundo: int, coluna_nova: str):
    """ conta os numeros das bolas entre 1 e 5, 11 e 20 e etc.. """
    contador = 0
    lista_numeros = []
    for jogo in df['jogos_ordenados']:
        for bola in jogo:
            if bola >= primeiro and bola <= segundo:
                contador += 1
        lista_numeros.append(contador)
        contador = 0
    df[coluna_nova] = lista_numeros
    

def contar_bolas(df: pd.DataFrame, qtd_jogos):
    """ conta os numeros de todas as bolas por jogo e faz a coluna """
    for bola in range(1, 26):
        contador = 0
        lista_numeros = [0 for i in range(len(df) - qtd_jogos)]
        for jogo in df.iloc[-qtd_jogos:, :]['jogos_ordenados']:
            if bola in jogo:
                contador += 1
            lista_numeros.append(contador)
        coluna_nova = "bola_" + str(bola) + "_sorteio" 
        df[coluna_nova] = lista_numeros


# Numero de sorteios em que a bola ficou sem sair
def sorteios_sem_sair_bola(df: pd.DataFrame):
    """ conta o numero de jogos que as bolas ficaram sem ser sorteadas """
    for bola in range(1, 26):
        jogo = 0
        soma = 0
        lista_numeros = []
        for linha in range(len(df['jogos_ordenados'])):
            if bola in df['jogos_ordenados'][linha]:
                jogo = linha
            soma = df['Concurso'].iloc[linha] - df['Concurso'].iloc[jogo] 
            lista_numeros.append(soma)
        coluna_nova = "bola_" + str(bola) + "_jogos_sem_sorteio" 
        df[coluna_nova] = lista_numeros
    

# Coloca em uma lista os numeros de sorteios em que a bola ficou sem sair
def resultado_sorteios_sem_sair_bola(df: pd.DataFrame) -> list:
    """ adiciona a qtd de sorteios que cada bola ficou sem ser sorteada em uma lista """
    lista_jogos = [0]
    for linha in range(1, len(df['jogos_ordenados'])):    
        lista_bolas = []
        for bola in range(1, 26):
            if bola in df['jogos_ordenados'][linha]:
                lista_bolas.append(df["bola_" + str(bola) + "_jogos_sem_sorteio"].iloc[linha-1]) 
        lista_jogos.append(lista_bolas)
    coluna_nova = 'sorteio_sem_sair_resultados'
    #print(lista_jo)
    df[coluna_nova] = lista_jogos


# Faz estatisticas de soma, media e desvio padrao das bolas que não saíram nos ultimos jogos
def soma_jogos_sem_sair(lista: list) -> int:
    """ retorna soma dos jogos """
    if lista == 0:
        pass
    else:
        return sum(lista)

def media_jogos_sem_sair(lista: list) -> int:
    """ retorna media dos jogos """
    if lista == 0:
        pass
    else:
        return round(np.mean(lista), 2)

def desvio_padrao_jogos_sem_sair(lista: list) -> int:
    """ retorna desvio padrao dos jogos """
    if lista == 0:
        pass
    else:
        return round(np.std(lista), 2)

# Quantidade de Numeros Primos,  de Fibonnacci, de Primo com Fibonnaci por Jogo, Pares, Impares e Divisíveis por 3.
def numeros_primos(df: pd.DataFrame):
    """ faz uma coluna com a quantidade de numeros primos por jogo"""
    primos = [1,2,3,5,7,11,13,17,19,23]
    qtd_primos = []
    for jogo in df['jogos_ordenados']:
        contador = 0
        for bola in jogo:
            if bola in primos:
                contador += 1
        qtd_primos.append(contador)
    df['qtd_primos'] = qtd_primos


def numeros_fibonacci(df: pd.DataFrame):
    """ faz uma coluna com a quantidade de numeros de fibonacci por jogo"""
    fibonacci = [1, 2, 3, 5, 8, 13, 21]
    qtd_fibonacci = []
    for jogo in df['jogos_ordenados']:
        contador = 0
        for bola in jogo:
            if bola in fibonacci:
                contador += 1
        qtd_fibonacci.append(contador)
    df['qtd_fibonacci'] = qtd_fibonacci


def numeros_fibonacci_e_primo(df: pd.DataFrame):
    """ faz uma coluna com a quantidade de numeros de fibonacci e primos por jogo"""
    fibonacci_primo = [1, 2, 3, 5,13]
    qtd_fibonacci_primo = []
    for jogo in df['jogos_ordenados']:
        contador = 0
        for bola in jogo:
            if bola in fibonacci_primo:
                contador += 1
        qtd_fibonacci_primo.append(contador)
    df['qtd_fibonacci_primo'] = qtd_fibonacci_primo


def pares_impares(df: pd.DataFrame):
    """ cria 2 colunas com as quantidades de numeros pares e impares por jogo"""
    pares = []
    impares = []
    for jogo in df['jogos_ordenados']:
        contador_par = 0
        contador_impar = 0
        for bola in jogo:
            if bola % 2 == 0:
                contador_par += 1
            else:
                contador_impar += 1
        pares.append(contador_par)
        impares.append(contador_impar)
    df['qtd_numeros_pares'] = pares
    df['qtd_numeros_impares'] = impares


def divisivel_por_tres(df: pd.DataFrame):
    """ cria coluna com a quantidade de numeros divisiveis por 3"""
    divisivel_tres = []
    for jogo in df['jogos_ordenados']:
        contador = 0
        for bola in jogo:
            if bola % 3 == 0:
                contador += 1
        divisivel_tres.append(contador)
    df['qtd_divisivel_tres'] = divisivel_tres


# Agrega todas as estatísticas anteriores em uma coluna só
def resultado_jogo(soma: int, media: int, desvio_padrao: int, entre_1_5: int, entre_6_10: int, 
                   entre_11_15: int, entre_16_20: int, entre_21_25: int, 
                   primos: int, fibonacci: int, primos_fibo: int, pares: int, impares: int, divisivel_3: int,
                   sem_sair_sum: int):
    
    return ([soma, media, desvio_padrao, entre_1_5, entre_6_10, entre_11_15, entre_16_20, entre_21_25,
             primos, fibonacci, primos_fibo, pares, impares, divisivel_3, sem_sair_sum])

# ------------------------- calls
# faz a lista dos jogos
lotofacil['jogos_ordenados'] = list(map(jogos_ordenados, lotofacil['bola_1'], lotofacil['bola_2'], lotofacil['bola_3'],
                              lotofacil['bola_4'], lotofacil['bola_5'], lotofacil['bola_6'],
                              lotofacil['bola_7'], lotofacil['bola_8'], lotofacil['bola_9'],
                              lotofacil['bola_10'], lotofacil['bola_11'], lotofacil['bola_12'],
                              lotofacil['bola_13'], lotofacil['bola_14'], lotofacil['bola_15']))

lotofacil['jogos_desordenados'] = list(map(jogos_desordenados, lotofacil['bola_1'], lotofacil['bola_2'], lotofacil['bola_3'],
                              lotofacil['bola_4'], lotofacil['bola_5'], lotofacil['bola_6'],
                              lotofacil['bola_7'], lotofacil['bola_8'], lotofacil['bola_9'],
                              lotofacil['bola_10'], lotofacil['bola_11'], lotofacil['bola_12'],
                              lotofacil['bola_13'], lotofacil['bola_14'], lotofacil['bola_15'])) 

# faz a soma dos jogos e deixa um inteiro
lotofacil['soma_jogos'] = list(map(soma_jogos, lotofacil['jogos_ordenados']))
# faz a média dos jogos e deixa um inteiro
lotofacil['media_jogos'] = list(map(media_jogos, lotofacil['jogos_ordenados']))
# faz o desvio padrão dos jogos
lotofacil['desvio_padrao_jogos'] = list(map(desvio_padrao_jogos, lotofacil['jogos_ordenados']))

lotofacil = lotofacil.drop(['bola_1', 'bola_2', 'bola_3', 'bola_4', 'bola_5', 'bola_6', 'bola_7',
                            'bola_8', 'bola_9', 'bola_10', 'bola_11', 'bola_12',
                            'bola_13', 'bola_14', 'bola_15'], axis=1)
entre_numeros(lotofacil, 1, 5, "entre_1_e_5")
entre_numeros(lotofacil, 6, 10, "entre_6_e_10")
entre_numeros(lotofacil, 11, 15, "entre_11_e_15")
entre_numeros(lotofacil, 16, 20, "entre_16_e_20")
entre_numeros(lotofacil, 21, 25, "entre_21_e_25") 

# Quantas vezes cada bola saiu nos ultimos 300 jogos
qtd_jogos = 300

contar_bolas(lotofacil, qtd_jogos)
sorteios_sem_sair_bola(lotofacil)
resultado_sorteios_sem_sair_bola(lotofacil)
lotofacil['sorteio_sem_sair_resultados_sum'] = list(map(soma_jogos_sem_sair, 
                                                        lotofacil['sorteio_sem_sair_resultados']))
lotofacil['sorteio_sem_sair_resultados_mean'] = list(map(media_jogos_sem_sair, 
                                                        lotofacil['sorteio_sem_sair_resultados']))
lotofacil['sorteio_sem_sair_resultados_std'] = list(map(desvio_padrao_jogos_sem_sair, 
                                                        lotofacil['sorteio_sem_sair_resultados']))
numeros_primos(lotofacil)
numeros_fibonacci(lotofacil)
numeros_fibonacci_e_primo(lotofacil)
pares_impares(lotofacil)
divisivel_por_tres(lotofacil)

lotofacil['resultados'] = list(map(resultado_jogo, lotofacil['soma_jogos'], lotofacil['media_jogos'], 
                                   lotofacil['desvio_padrao_jogos'],
                              lotofacil['entre_1_e_5'], lotofacil['entre_6_e_10'], lotofacil['entre_11_e_15'],
                              lotofacil['entre_16_e_20'], lotofacil['entre_21_e_25'], lotofacil['qtd_primos'],
                              lotofacil['qtd_fibonacci'], lotofacil['qtd_fibonacci_primo'],
                                   lotofacil['qtd_numeros_pares'],
                              lotofacil['qtd_numeros_impares'], lotofacil['qtd_divisivel_tres'], 
                                   lotofacil['sorteio_sem_sair_resultados_sum']))
```

```python
import random

def pick_random_right_game(resultado, vinte_cinco_primeiros_list, first_row, 
                           second_row, third_row, fourth_row, fifth_row):
    counter_combinations = 0
    first = 0
    while(first != resultado):    
        counter_combinations += 1
        first = random.sample(vinte_cinco_primeiros_list[0:5], first_row)
        first = [x[0] for x in first]

        second = random.sample(vinte_cinco_primeiros_list[5:10], second_row)
        second = [x[0] for x in second]

        third = random.sample(vinte_cinco_primeiros_list[10:15], third_row)
        third = [x[0] for x in third]

        fourth = random.sample(vinte_cinco_primeiros_list[15:20], fourth_row)
        fourth = [x[0] for x in fourth]

        fifth = random.sample(vinte_cinco_primeiros_list[20:25], fifth_row)
        fifth = [x[0] for x in fifth]

        first = first + second + third + fourth + fifth
        first = sorted(first)
    return [first, counter_combinations] 

lista_game_counter = pick_random_right_game(resultado, vinte_cinco_primeiros_list, 3, 4, 4, 2, 2)
print(lista_game_counter[0])
print(lista_game_counter[1])
```

```python
%%time
# 0 soma (170 - 189 e 200 - 220)
# 1 media (11 - 14)
# 2 desvio padrao (6, 7 e 8)
# 3 entre numeros 1 e 5 (2, 3 e 4)
# 4 entre numeros 6 e 10 (2, 3, 4, 5)
# 5 entre numeros 11 e 15 (2, 3, 4)
# 6 entre numeros 16 e 20 (2, 3, 4)
# 7 entre numeros 21 e 25 (2, 3, 4)
# 8 primos (4, 5, 6, 7)
# 9 fibonacci (3,4,5)
# 10 primos e fibo (2, 3, 4)
# 11 pares (6, 7, 8)
# 12 impares (7 , 8 , 9)
# 13 divisivel_3 (3, 4, 5, 6)
# 14 sem sair 16 dias de sum

def get_games(df: pd.DataFrame) -> list:
    index = df.index
    sem_16 = []
    soma = []
    mean = []
    std = []
    one_five = []
    six_ten = []
    eleven_fifteen = []
    sixteen_twenty = []
    tweone_twefive = []
    prime = []
    fibo = []
    prime_fibo = []
    odds = []
    even = []
    div_3 = []
    final = []
    
        
    for sorteios in index:
        if (df['resultados'][sorteios][14] == 19):
            sem_16.append(sorteios)
    for primo_fibo in sem_16:
        if (df['resultados'][primo_fibo][10] >= 2) and (df['resultados'][primo_fibo][10] <= 4):
            div_3.append(primo_fibo)
    for n in div_3:
        if (df['resultados'][n][13] >= 3) and (df['resultados'][n][13] <= 6):
            one_five.append(n)
    for n in one_five:
        if (df['resultados'][n][3] >= 2) and (df['resultados'][n][3] <= 4):
            soma.append(n)
    for n in soma:
        if ((df['resultados'][n][0] >= 170) and (df['resultados'][n][0] <= 189) or \
            (df['resultados'][n][0] >= 200) and (df['resultados'][n][0] <= 220)):
            mean.append(n)
    for n in mean:
        if (df['resultados'][n][1] >= 11) and (df['resultados'][n][1] <= 14):
            std.append(n)
    for n in std:
        if (df['resultados'][n][2] >= 6) and (df['resultados'][n][2] <= 8):
            six_ten.append(n)
    for n in six_ten:
        if (df['resultados'][n][4] >= 2) and (df['resultados'][n][4] <= 5):
            eleven_fifteen.append(n)
    for n in eleven_fifteen:
        if (df['resultados'][n][5] >= 2) and (df['resultados'][n][5] <= 4):
            sixteen_twenty.append(n)
    for n in sixteen_twenty:
        if (df['resultados'][n][6] >= 2) and (df['resultados'][n][6] <= 4):
            tweone_twefive.append(n)
    for n in tweone_twefive:
        if (df['resultados'][n][7] >= 2) and (df['resultados'][n][7] <= 4):
            prime.append(n)
    for n in prime:
        if (df['resultados'][n][8] >= 4) and (df['resultados'][n][8] <= 7):
            fibo.append(n)
    for n in fibo:
        if (df['resultados'][n][9] >= 3) and (df['resultados'][n][9] <= 5):
            odds.append(n)
    for n in odds:
        if (df['resultados'][n][11] >= 6) and (df['resultados'][n][11] <= 8):
            even.append(n)
    for n in even:
        if (df['resultados'][n][12] >= 7) and (df['resultados'][n][12] <= 9):
            final.append(n)

    return final

lista_final_2 = get_games(data_jogos)
```
